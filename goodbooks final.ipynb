```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Purvitha S\OneDrive\good books\books.csv")
```

```python
df = df.drop(columns=['isbn', 'isbn13', 'image_url', 'small_image_url'], errors='ignore')
```

```python
df.isnull().sum()
```

```python
df['books_count'] = pd.to_datetime(df['books_count'], dayfirst=True)
```

```python
df['Month'] = df['books_count'].dt.to_period('M')
df['Month']
```

```python
df.describe()
```

```python

avg_rating_by_author = df.groupby('authors')['average_rating'].mean().sort_values(ascending=False)
print("\nAverage Rating by Author:")
print(avg_rating_by_author.head(10))

ratings_count_by_author = df.groupby('authors')['ratings_count'].sum().sort_values(ascending=False)
print("\nTotal Ratings by Author:")
print(ratings_count_by_author.head(10))
```

```python

correlation_matrix = df[['ratings_count', 'average_rating']].corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
df = pd.read_csv(r"C:\Users\Purvitha S\OneDrive\good books\books.csv")
df = df.dropna(subset=['authors', 'ratings_count', 'average_rating'])

print("\nAverage Rating by Author:")
print(df.groupby('authors')['average_rating'].mean().sort_values(ascending=False).head(10))

print("\nAverage Ratings Count by Language Code:")
print(df.groupby('language_code')['ratings_count'].mean().sort_values(ascending=False).head(10))

if 'publisher' in df.columns:
    print("\nAverage Rating by Publisher:")
    print(df.groupby('publisher')['average_rating'].mean().sort_values(ascending=False).head(10))
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8, 5))
df.groupby('authors')['average_rating'].mean().sort_values(ascending=False).head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Authors by Average Rating')
plt.ylabel('Average Rating')
plt.xlabel('Author')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['original_publication_year'].value_counts().sort_index().plot(kind='line', marker='o', color='green')
plt.title('Books Published per Year')
plt.ylabel('Number of Books')
plt.xlabel('Publication Year')
plt.tight_layout()
plt.show()
```

```python
df.nlargest(10, 'ratings_count').set_index('original_title')['ratings_count'].plot(kind='barh', color='orange')
plt.title('Top 10 Most Rated Books')
plt.xlabel('Ratings Count')
plt.ylabel('Book Title')
plt.tight_layout()
plt.show()
```

```python
df.plot(kind='scatter', x='ratings_count', y='average_rating', alpha=0.5)
plt.title('Ratings Count vs Average Rating')
plt.xlabel('Ratings Count')
plt.ylabel('Average Rating')
plt.tight_layout()
plt.show()
```

```python
df['average_rating'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number').columns

plt.figure(figsize=(10,6))
sns.boxplot(data=df[numeric_cols])
plt.title('Box Plots for Numeric Columns')
plt.xticks(rotation=45)
plt.show()
```
